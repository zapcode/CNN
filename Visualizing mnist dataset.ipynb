```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import tensorflow as tf
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' #Removing Unwanted Logs
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data',one_hot= True)
```

```python
x = mnist.train.images.shape
y = mnist.test.images.shape
z = mnist.train.labels.shape
w = mnist.test.labels.shape
print(x)
print(y)
print(z)
print(w)
```

```python
with tf.Session() as sess:
    for i in range(2):
        first_image = mnist.train.images[i]
        first_image = np.array(first_image, dtype='float')
        pixels = first_image.reshape((28, 28))
        plt.imshow(pixels, cmap='gray_r')  #cmap = 'gray or 'gray_r
        plt.title("True : %i "%(sess.run(tf.argmax(mnist.train.labels[i]))))
        plt.show()
```
